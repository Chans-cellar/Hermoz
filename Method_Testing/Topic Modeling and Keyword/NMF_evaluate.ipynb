```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
# Load the dataset
df = pd.read_csv('E:\\studies\\USJ FOT\\lecture\\Research\\CodeBase\\CreatingDataset\\cleaned\\acroeconomic_classifier_dataset_4_hypercleaned_with_nmf_topics.csv')
```

```python
# Define the order of your labels
label_order = ['Inflation', 'International Trade', 'GDP Growth', 'Exchange Rates', 'Monetary Policy', 'Fiscal Policy', 'Unemployment']

# Create a mapping from labels to integers
label_to_int = {label: idx for idx, label in enumerate(label_order)}

# Convert the labels to integers
df['label_int'] = df['label'].map(label_to_int)
```

```python
# Create the confusion matrix
conf_matrix = confusion_matrix(df['label_int'], df['nmf_topic'])

# Plot the confusion matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=range(7), yticklabels=label_order)
plt.xlabel('NMF Topics')
plt.ylabel('True Labels')
plt.title('Confusion Matrix of True Labels vs. NMF Topics')
plt.show()
```
